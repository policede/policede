```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bsp
import requests
import matplotlib.pyplot as plt
%matplotlib inline


hitter_data = pd.read_excel('1982_2017_hitter_half_money_start.xlsx')
hitter_data2 =  hitter_data[hitter_data['season'] ==  2016]
hitter_data2 =  hitter_data2[hitter_data2['salary'] != 'retire']

namecode_list=[]

for i in hitter_data2['URL']:
    namecode_list.append(i[-5:-1]+i[-1])
        
hitter_data2['namecode'] = namecode_list

salary_list=[]
for i in hitter_data2['salary']:
    salary_list.append(int(i[:-2]))
    
hitter_data2['salary'] = salary_list

startyear_list =[]
for i in hitter_data2['start']:
    if int(i[0]) == 9:
        j = str(19)+ i
        startyear_list.append(j[0:4])
    if int(i[0]) != 9:
        j = str(20) + i
        startyear_list.append(j[0:4])
hitter_data2['start'] = startyear_list


season_list = []
for i in hitter_data2['season']:
    i = str(i)
    season_list.append(i)

hitter_data2['season'] = season_list

year_list=[]
for i in range(len(hitter_data2['season'])):
    
    j = int(season_list[i]) - int(startyear_list[i]) +1
    year_list.append(j)
    
    
hitter_data2['year'] = year_list
```

```python
import pandas as pd
import statsmodels.formula.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
hitter_data2
result = sm.ols(formula = 'salary ~ WAR', data = hitter_data2).fit()
print (result.summary())
print('Predicted values',  result.predict())
result.pvalues

fig = plt.figure(figsize=(10,10))
plt.scatter(hitter_data2['WAR'], hitter_data2['salary'],s=90)
plt.tick_params(labelsize=10)
plt.plot(hitter_data2['WAR'],  result.predict(), 'r-', label="regression")
plt.xlabel("WAR")
plt.ylabel("salary")
plt.title("salary and WAR")
plt.legend()
plt.show()
```

```python
result = sm.ols(formula = 'salary ~ year', data = hitter_data2).fit()
print (result.summary())
print('Predicted values',  result.predict())
result.pvalues

fig = plt.figure(figsize=(10,10))
plt.scatter(hitter_data2['year'], hitter_data2['salary'],s=90)
plt.tick_params(labelsize=10)
plt.plot(hitter_data2['year'],  result.predict(), 'r-', label="regression")
plt.xlabel("year")
plt.ylabel("salary")
plt.title("salary and year")
plt.legend()
plt.show()
```
